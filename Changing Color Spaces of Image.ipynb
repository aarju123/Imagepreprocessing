# Convert image color

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("tajmahal.jpg")
```

```python
plt.imshow(image)
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
```
